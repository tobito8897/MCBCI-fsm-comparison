<h3><b>Introducción</b></h3>
<p style="text-align: justify; font-size:14px">De acuerdo con Kammerman & Wasserman (2001), “la epilepsia era el tercer desorden neurológico con mayor grado de incidencia, después de los infartos cerebrales y la enfermedad de Alzheimer”. Además, la OMS (2019) estima que “en todo el planeta hay aproximadamente 50 millones de personas con epilepsia. La proporción de la población con epilepsia (activa en algún momento de su vida) oscilaría de 4 a 10 por 1000 personas” sin embargo, estos valores podrían alcanzar los 14 por 1000 habitantes en países de ingreso mediano y bajo.</p>
<p style="text-align: justify; font-size:14px">La detección y el análisis de crisis epilépticas en registros de electroencefalografía, resulta ser una tarea primordial para darle un tratamiento y un seguimiento adecuado a pacientes con este padecimiento. Como afirma Schachter et al. (2013), un diagnóstico preciso permite seleccionar el tratamiento más apropiado, identificar efectos secundarios, reconocer patrones y posibles agentes desencadenantes.</p>
<p style="text-align: justify; font-size:14px">Diversos algoritmos de aprendizaje automático se han planteado para la detección de epilepsia (clasificación automática). Para el entrenamiento de estos modelos se han usado características muy diversas, las cuales son calculadas a partir del electroencefalograma. Por mencionar algunas: promedio, varianza, curtosis y entropía. Algunos analizan la señal en el dominio del tiempo, otros caracterizan su espectro de potencia y otros más, usan Wavelet para un análisis tiempo-frecuencia (Siddiqui et al., 2020).</p>
<p style="text-align: justify; font-size:14px">La selección de <i>features</i> para modelos de clasificacion de epilepsia ya ha sido abordada con anterioridad. Fergus et al. (2016) calcularon valores estadísticos a partir de las bandas del EEG, mediante el uso de algoritmos como Análisis de Compontes Principales y Análisis del Discriminante Lineal, seleccionaron el top 20 de características. Khati & Ingle (2020) usaron la validación cruzada para disminuir el número de atributos de 11 a 8, obteniendo una precisión del 99% al usar la regresión logística. Lai et al. (2017), usaron imágenes de resonancia para la clasificación de pacientes con epilepsia en el lóbulo temporal, la mejor precisión (84%) se obtuvo al usar una máquina de soporte vectorial para seleccionar los ragos de mayor importancia.</p>
</body>

<h3><b>Planteamiento del problema</b></h3>
<p style="text-align: justify; font-size:14px">Dada la diversidad de medidas que se pueden obtener del electroencefalograma, podría es complicado seleccionar las características más adecuadas para generar un modelo de clasificación automática.</p>
</body>

<body>
<h3><b>Justificación</b></h3>
<p style="text-align: justify; font-size:14px">Un paso en el diseño de cualquier modelo es la selección de características o atributos, entendiendo características como medidas de las propiedades del fenómeno de estudio. En muchos casos estas pueden ser seleccionadas de acuerdo al objetivo de nuestro modelo.</p>
<p style="text-align: justify; font-size:14px">En el caso de los modelos de clasificación automática el proceso de selección es de suma importancia, porque los atributos elegidos deben de reflejar las diferencias entre cada categoría.</p>
<p style="text-align: justify; font-size:14px">A partir del electroencefalograma se pueden calcular un sinnúmero de atributos. Si bien es posible introducir todos estos valores al algoritmo de inteligencia artificial, en ocasiones las características que no aportan a la clasificación pueden empeorar el rendimiento del modelo. Además un modelo más complejo tardará más en ser entrenado.</p>
<p style="text-align: justify; font-size:14px">Si bien hay trabajos previos acerca de la clasificación de señaes ictales y no ictales, considero que es necesario establecer procesos que nos permitan optimizar los algoritmos mediante una adecuada selección de características.</p>
</body>

<h4><b>Cálculo de características</b></h4>
<p style="text-align: justify; font-size:14px">Las señales de electroencefalografía estaban dadas en volts; se convirtieron a microvoltios, para no manejar valores decimales muy pequeños.<p>
<p style="text-align: justify; font-size:14px">Se uso un filtro pasa-bajas tipo <i>butterworth</i>, con una frecuencia de corte de 0.5Hz y orden 2. Las frecuencias bajas se asocian con movimientos del paciente durante el registro.<p>
<p style="text-align: justify; font-size:14px">Sigiendo un procedimiento similar a Fergus et al. (2016), la señal fue segmentada en 5 bandas:<p>
    <ul>
    <li style="font-size:14px">Rango de frecuencias completo: 0.5–30 Hz</li>
    <li style="font-size:14px">Delta: 0.5 < f < 4 Hz</li>
    <li style="font-size:14px">Theta: 4 < f < 8 Hz</li>
    <li style="font-size:14px">Alfa: 8 < f < 12 Hz</li>
    <li style="font-size:14px">Beta: 12 < f < 25 Hz</li>
    </ul>
<p style="text-align: justify; font-size:14px">Para calcular las bandas se usaron filtros pasa-bandas de tipo <i>butterworth</i>, orden 2.<p>
<p style="text-align: justify; font-size:14px">Para cada banda y canal, se calculó:<p>
    <ul>
    <li style="font-size:14px">Frecuencia pico</li>
    <li style="font-size:14px">Mediana de la frecuencia</li>
    <li style="font-size:14px">Varianza</li>
    <li style="font-size:14px">Valor RMS</li>
    <li style="font-size:14px">Asimetría</li>
    <li style="font-size:14px">Curtosis</li>
    <li style="font-size:14px">Desviación Estándar</li>
    <li style="font-size:14px">Cruces por cero</li>
    <li style="font-size:14px">Entropía Muestral</li>
    </ul>
<p style="text-align: justify; font-size:14px">Luego del proceso anterior, para cada ventana se obtuvo un  vector de longitud: (5 bandas X 22 canales X 6 características).<p>

<h4><b>Diseño de la red neuronal</b></h4>
<p style="text-align: justify; font-size:14px">Se usó un diseño con 2 capas intermedias de  100 y 50 neuronas, y una capa de salida de 2 neuronas.<p>
<p style="text-align: justify; font-size:14px">Las capas intermedias usaron una activación de tipo Relu y la de salida, una sigmoide. Se utilizó el optimizador Adam y la entropía cruzada binaria como medida de error.<p>

```python
top_features = 990
ANN = Net(top_features)
```

<h4><b>Entrenamiento de la Red Neuronal</b></h4>
<p style="text-align: justify; font-size:14px">Para entrenar la red se utilizó una validación cruzada de 3 iteraciones, 66% de datos para entrenamiento y 33% para evaluación.<p>
<p style="text-align: justify; font-size:14px">La red neuronal fue entrenada durante 200 épocas y 5 pasos por época. Si la entropía cruzada no disminuia durante 20 épocas, el entrenamiento se detenía.<p>
<p style="text-align: justify; font-size:14px">Posterior al entrenamiento, el modelo era evaluado con los datos de prueba (33%).<p>

```python
ictal_path = "../ProcessedData/Ictal/"
noictal_path = "../ProcessedData/NoIctal/"
features_map_file = "../HelperFiles/features_map.json"
output_directory = "../MLData/"

explainer = "tree" 
model = "ann"
top_dir = output_directory + "/TopFeatures/" + "/top_%s_%s.pickle" % (model, explainer)
importances = read_pickle(top_dir)

features_map = get_json_content(features_map_file)
data, labels = prepare_data_for_ml(ictal_path, noictal_path)

k_folds_stratified_nn(model, folds, data, labels, output_directory,
                      features_map, 990, explainer, importances, start_kwargs[model],
                      train_kwargs[model])
```

<h4><b>Eliminación recursiva de características</b></h4>
<p style="text-align: justify; font-size:14px">Inspirado por Khati & Ingle (2020). Se realizó una eliminación de características de acuerdo a su importancia, pero en este caso se usó el método recursivo.<p>
<p style="text-align: justify; font-size:14px">Para obtener la importancia de cada característica se usó un Árbol de Decisión, se utilizó el valor de impureza de Gini. La profundidad máxima era cuando el  nodo contenía una muestra.<p>
<p style="text-align: justify; font-size:14px">La importancia se calcula como el decremento en la impureza de la entrada de un nodo y sus salidas, luego se pondera por el número de muestras del nodo. Este valor se promedia por características y para todos los nodos del árbol (Lee, 2020).<p>
<p style="text-align: justify; font-size:14px">La red neuronal fue entrenada y evaluada con una cantidad diferente de características (650 a 1 característica en saltos de 50). Siempre se mantuvieron las características con mayor importancia.<p>

<h3><b>Discusión</b></h3>
    <p style="text-align: justify; font-size:14px">A partir del <i>ranking</i> calculado por el árbol de decisiones, podemos observar que muchas características calculadas en la señal completa (0.5-30Hz) tiene poca relevancia para la clasificación. Es posible que diversos rasgos que diferencian una señal ictal, de una no ictal, solo sean visibles al realizar un análisis por banda.</p>
        <p style="text-align: justify; font-size:14px">La asimetría y la frecuencia pico no aparecen en el <i>top</i> 20, por lo cual podriamos intuir que estos parámetros son similares en registros de EEG de personas sanas y con epilepsia.</p>
        <p style="text-align: justify; font-size:14px">Es necesario tomar en cuenta que la base de datos no menciona el tipo de epilepsia. Si las epilepsias fueran de tipo focal, la actividad ictal sería más evidente en un grupo de canales. Lo anterior generaría que a las características estimadas para estos canales se les asignara mayor importancia. El <i>ranking</i> del árbol de decisiones no muestra que exista predominancia de algunos canales específicos.</p>
<p style="text-align: justify; font-size:14px">El desempeño de la red neuronal (0.888) es menor que el reportado por otros modelos de inteligencia artificial. En el presente trabajo no se evaluaron otros algoritmos de aprendizaje automático, considero que mediante la evaluación de algoritmos alternativos y el ajuste de parámetros es posible obtener una mejor precisión.</p>
<p style="text-align: justify; font-size:14px">La mayor precisión (0.888) del modelo se obtuvo al usar 250 características para entrenar el modelo. Se observa que al disminuir la cantidad de características, el desempeño no disminuye de manera notable. Al usar solo las 50 características más importantes, la precisión es de 0.873. Cuando se usó 1 característica, el decremento si fue notorio (0.746).</p>
<p style="text-align: justify; font-size:14px">Podemos confirmar que la eliminación recursiva de características es útil para simplificar nuestro modelo, sin que la simplificación tenga un impacto en el desempeño. Además, el <i>ranking</i> mediante el árbol de decisiones permite que los resultados sean facilmente interpretables, a diferencia de otros métodos como el Análisis de Componentes Principales, en los cuales se generan combinaciones lineales de los valores.</p>
